```python
def customplot(H,sigma,EYE,t):
#     X1,Y1,Z1=X1,Y1,Z1
#     X2,Y2,Z2=X2,Y2,Z2
    
    m1=np.cross(H,[0,0,1])/np.linalg.norm(np.cross(H,[0,0,1]))
    [X1,Y1,Z1]=m1
    fig1 = go.FigureWidget()
    #NEEL VECTOR1
    x1 = np.array([0,X1])
    y1 = np.array([0,Y1])
    z1 = np.array([0,Z1])

    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.75,
        marker=dict(size=0.1,color='black'),
        line=dict(color='red', width=4 ),
        name="$m_{1}$")
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]], u=[x1[1]+x1[1]], v=[y1[1]+y1[1]], w=[z1[1]+z1[1]],opacity=0.75,showscale=False,anchor="tip",sizemode="absolute",
        sizeref=0.1,colorscale=[[0, 'red'], [1,'red']])

    m2=-1*m1
    [X2,Y2,Z2]=m2
    #NEEL VECTOR2
    x2 = np.array([0,X2])
    y2 = np.array([0,Y2])
    z2 = np.array([0,Z2])
    fig1.add_scatter3d(x=x2, y=y2, z=z2,opacity=0.75,
        marker=dict(size=0.1,color='black'),
        line=dict(color='blue', width=4 ),
        name="$m_{2}$")
    fig1.add_cone(x=[x2[1]], y=[y2[1]], z=z2[[1]], u=[x2[1]+x2[1]], v=[y2[1]+y2[1]], w=[z2[1]+z2[1]],opacity=0.75,showscale=False,anchor="tip",sizemode="absolute",
        sizeref=0.1,colorscale=[[0, 'blue'], [1,'blue']])
    
###################################
    ScaleTorques=0.5
    #Torques on Neel Vector1

    m1=[X1,Y1,Z1]
    FL1=np.cross(sigma,m1)
    FL1=FL1*ScaleTorques

    x1 = np.array([X1,X1+FL1[0]])
    y1 = np.array([Y1,Y1+FL1[1]])
    z1 = np.array([Z1,Z1+FL1[2]])

    #Field-like Torque
    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.9,
        marker=dict(size=0.1,color='black'),
        line=dict(color='orange', width=5 ),
        name="$\\tau_{FL}$",showlegend=True)
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]],u=[x1[1]-x1[0]], v=[y1[1]-y1[0]], w=[z1[1]-z1[0]],opacity=0.75,showscale=False,anchor="tail",sizemode="absolute",
        sizeref=0.05,colorscale=[[0, 'orange'], [1,'orange']])


    DL1=np.cross(m1,np.cross(sigma,m1))
    DL1=DL1*ScaleTorques
    x1 = np.array([X1,X1+DL1[0]])
    y1 = np.array([Y1,Y1+DL1[1]])
    z1 = np.array([Z1,Z1+DL1[2]])

    #Damping-like Torque
    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.9,
        marker=dict(size=0.1,color='black'),
        line=dict(color='brown', width=5 ),
        name="$ \\tau_{DL}$",showlegend=True)
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]], u=[x1[1]-x1[0]], v=[y1[1]-y1[0]], w=[z1[1]-z1[0]],opacity=0.75,showscale=False,anchor="tail",sizemode="absolute",
        sizeref=0.05,colorscale=[[0, 'brown'], [1,'brown']])

    #Torques on Neel Vector2

    m2=[X2,Y2,Z2]
    FL2=np.cross(sigma,m2)
    FL2=FL2*ScaleTorques

    x1 = np.array([X2,X2+FL2[0]])
    y1 = np.array([Y2,Y2+FL2[1]])
    z1 = np.array([Z2,Z2+FL2[2]])
    #Field-like Torque
    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.9,
        marker=dict(size=0.1,color='black'),
        line=dict(color='orange', width=5 ),
        name="$ \Tau_{FL}$",showlegend=False)
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]], u=[x1[1]-X2], v=[y1[1]-Y2], w=[z1[1]-Z2],opacity=0.75,showscale=False,anchor="tail",sizemode="absolute",
        sizeref=0.05,colorscale=[[0, 'orange'], [1,'orange']])
    


    DL2=np.cross(m2,np.cross(sigma,m2))
    DL2=DL2*ScaleTorques

    x1 = np.array([X2,X2+DL2[0]])
    y1 = np.array([Y2,Y2+DL2[1]])
    z1 = np.array([Z2,Z2+DL2[2]])

    #Damping-like Torque
    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.9,
        marker=dict(size=0.1,color='black'),
        line=dict(color='brown', width=5 ),
        name="$ \tau_{DL}$",showlegend=False)
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]], u=[x1[1]-X2], v=[y1[1]-Y2], w=[z1[1]-Z2],opacity=0.75,showscale=False,anchor="tail",sizemode="absolute",
        sizeref=0.05,colorscale=[[0, 'brown'], [1,'brown']])
##################################

    #excessmoment2
    #factor=10
    #x2 = factor*np.array([0,X1s[-1]+X2s[-1]])
    #y2 = factor*np.array([0,Y1s[-1]+Y2s[-1]])
    #z2 = factor*np.array([0,Z1s[-1]+Z2s[-1]])
    #fig1.add_scatter3d(x=x2, y=y2, z=z2,opacity=0.75,
    #    marker=dict(size=0.1,color='black'),
    #    line=dict(color='black', width=4 ),
    #    name="$M$")
    #fig1.add_cone(x=[x2[1]], y=[y2[1]], z=z2[[1]], u=[x2[1]+x2[1]], v=[y2[1]+y2[1]], w=[z2[1]+z2[1]],opacity=0.75,showscale=False,anchor="tip",sizemode="absolute",
    #    sizeref=0.01,colorscale=[[0, 'black'], [1,'black']])


    fig1.update_layout(
        height=700,
        width=900,

        autosize=True,
        scene=dict(camera=dict(up=dict(x=0,y=0, z=1 ), eye=dict(x=EYE[0],y=EYE[1],z=EYE[2])),
                   xaxis_title='',
                   yaxis_title='',
                   zaxis_title='',
                   xaxis = dict(nticks=3, range=[-1.25,1.25],showticklabels=False),
                   yaxis = dict(nticks=3, range=[-1.25,1.25],showticklabels=False),
                   zaxis = dict(nticks=3, range=[-1.25,1.25],showticklabels=False),
            aspectratio = dict( x=1, y=1, z=1),
            aspectmode = 'manual')
    )
    

    # UNIT SPHERE
    theta, phi = np.mgrid[0:2*np.pi:100j, 0:2*np.pi:100j]
    x = np.sin(theta)*np.sin(phi)
    y = np.sin(theta)*np.cos(phi)
    z = np.cos(theta)

    fig1.add_surface(x=x,y=y,z=z,opacity=0.07,colorscale=[[0, 'purple'], [1,'orange']],showlegend=False, showscale = False)

#     #Orbital Path1
#     fig1.add_scatter3d(x=X1, y=Y1, z=Z1s,opacity=0.25,
#         marker=dict(size=0.1,color='red'),
#         line=dict(color='red', width=4),
#         showlegend=False)
#     #Orbital Path1
#     fig1.add_scatter3d(x=X2s, y=Y2s, z=Z2s,opacity=0.25,
#         marker=dict(size=0.1,color='blue'),
#         line=dict(color='blue', width=4 ),
#         showlegend=False)

    #Extermal Field1
    [Hx,Hy,Hz]=H
    norm=np.linalg.norm([Hx,Hy,Hz])
    x1 = np.array([0,Hx])/(2*norm)
    y1 = np.array([0,Hy])/(2*norm)
    z1 = np.array([0,Hz])/(2*norm)

    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.75,
        marker=dict(size=0.1,color='black'),
        line=dict(color='green', width=6 ),
        name="$H_{applied}$")
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]], u=[x1[1]+x1[1]], v=[y1[1]+y1[1]], w=[z1[1]+z1[1]],opacity=0.75,showscale=False,anchor="cm",sizemode="absolute",
        sizeref=0.1,colorscale=[[0, 'green'], [1,'green']])

    # Spin Accumulation

    x1 = np.array([0,sigma[0]])
    y1 = np.array([0,sigma[1]])
    z1 = np.array([0,sigma[2]])

    fig1.add_scatter3d(x=x1, y=y1, z=z1,opacity=0.75,
        marker=dict(size=0.1,color='black'),
        line=dict(color='yellow', width=6),
        name="$\sigma_{AC}$")
    fig1.add_cone(x=[x1[1]], y=[y1[1]], z=z1[[1]], u=[x1[1]+x1[1]], v=[y1[1]+y1[1]], w=[z1[1]+z1[1]],opacity=0.75,showscale=False,anchor="tail",sizemode="absolute",
        sizeref=0.1,colorscale=[[0, 'yellow'], [1,'yellow']])

    #CIRCLE IN XY PLANE
    C1=1
    phi=np.linspace(0,2*np.pi,100)
    x=C1*np.cos(phi)
    y=C1*np.sin(phi)
    z=C1*np.zeros(len(x))
    fig1.add_scatter3d(x=x, y=y, z=z,opacity=0.2,
        marker=dict(size=0.1,color='grey'),
        line=dict(color='grey', width=2 ),
        showlegend=False)
    #CIRCLE IN XZ PLANE
    phi=np.linspace(0,2*np.pi,100)
    x=C1*np.cos(phi)
    y=C1*np.zeros(len(x))
    z=C1*np.sin(phi)
    fig1.add_scatter3d(x=x, y=y, z=z,opacity=0.2,
        marker=dict(size=0.1,color='grey'),
        line=dict(color='grey', width=2 ),
        showlegend=False)

    #CIRCLE IN YZ PLANE
    phi=np.linspace(0,2*np.pi,100)
    x=C1*np.zeros(len(x))
    y=C1*np.cos(phi)
    z=C1*np.sin(phi)
    fig1.add_scatter3d(x=x, y=y, z=z,opacity=0.2,
        marker=dict(size=0.1,color='grey'),
        line=dict(color='grey', width=2 ),
        showlegend=False)
    if t<10:
        name="000"+str(t)
    elif t<100:
        name="00"+str(t)
    elif t<1000:
        name="0"+str(t)
    elif t<10000:
        name=str(t)
    pio.write_image(fig1,r"C:\Users\Egecan\Desktop\XZ\\"+name+".png",height=1000,width=1200)

    
    
    
    
```

```python
customrange=np.arange(0,360,1)
R=1.75
aor=0*(np.pi/180)
EYES=[]
steps=len(customrange)
for index,angle in enumerate(customrange):
    EYE=R*np.asfarray([np.cos(np.pi/4+index*aor/steps),np.sin(np.pi/4+index*aor/steps),np.cos(np.pi/3)])
    Hscan=[np.cos((angle/180)*np.pi),0,np.sin((angle/180)*np.pi)]
    sigma=[0,np.cos(12*(angle/180)*np.pi)/2,0]
    customplot(Hscan,sigma,EYE,angle)


    
```
